# Dealing With Missing Values # 

```python
# checking missing value
raw_data = pd.read_csv('sensor_raw_data.csv', sep = '\t')
```

```python
raw_data.isnull().sum()
```

```python
# Removing 9419 rows, given that 9419 appear so many times, indicating that these rows do not provide information
remove_rows = raw_data['event'].isnull()
raw_data_removed = raw_data[:][~remove_rows]
raw_data_removed.shape
```

```python
# checking more missing values to see if certain columns should be removed
raw_data_removed.isnull().sum()
```

```python
# Removing columns that have 65673 missing values (providing no information)
cleaned_data = raw_data_removed.drop(['first_browser_language'], axis=1)
cleaned_data = cleaned_data.drop(['first_referrer'], axis=1)
cleaned_data = cleaned_data.drop(['first_referrer_host'], axis=1)
cleaned_data = cleaned_data.drop(['first_visit_time'], axis=1)
cleaned_data.shape
```

```python
cleaned_data_csv = 'Documents/DS501/Data/sensor_cleaned_data.csv'
cleaned_data.to_csv('sensor_cleaned_data.csv', sep='\t', encoding='utf-8', index = False)
```
